# El problema de la mochila

Se tiene una mochila y un conjunto de artículos con diferente peso y valor. Se intenta determinar que artículos se pueden colocar adentro de la mochila de manera que no se sobrepase el límite de peso de la mochila maximizando el valor total. 

## Primera parte

Resolver el problema utilizando un método de fuerza bruta para los siguientes artículos y un límite de 15kg:

<table style="font-size:16px">
    <tr>
        <th>Item</th><td>Valor</td><td>Peso</td>
    </tr>
    <tr>
        <th>A</th><td>4</td><td>12</td>
    </tr>
    <tr>
        <th>B</th><td>2</td><td>2</td>
    </tr>
    <tr>
        <th>C</th><td>10</td><td>4</td>
    </tr>
    <tr>
        <th>D</th><td>1</td><td>1</td>
    </tr>
</table>

Mostrar el resultado para todos los casos

```python
import itertools

maximo_valor = 0
peso_maximo = 15
combination = []
items = [(4, 12),(2, 2),(10, 4),(1, 1)]

def Solution(item_seleccionado, maximo_valor):    
    print(f'\n(Valor, Peso) = {item_seleccionado})')
    print(f'Valor total: {maximo_valor}')

def main(maximo_valor, peso_maximo, combination, items):

    for i in range(1, 5):
        combination.append([i for i in itertools.combinations(items, i)])

    for item in combination:
        for j in item:
            total_valor = 0
            peso = 0
            for i in j:
                valor, partial_peso = i
                total_valor += valor
                peso += partial_peso
            if peso <= peso_maximo and maximo_valor < total_valor:
                item_seleccionado = j
                maximo_valor = total_valor
            print(f"Peso: {peso} | Valor: {total_valor}")
            
    Solution(item_seleccionado, maximo_valor)

if __name__ == '__main__':
    main(maximo_valor, peso_maximo, combination, items)
    
```

```python
data = [
        (4, 12),(2, 2),
        (10, 4),(1, 1),
        (5, 15),(3, 2),
        (14, 7),(4, 10)]

history = []
history_extract = []
weight = 0
go_out = False

def extract_history(history_extract):
    print(history_extract)
    
def main(data,history,history_extract,weight,go_out):
    
    data.sort()
    metric = [float(k / v) for k, v in data]
    print(metric)

    while True:
        max_metric = max(metric, key=float)
        metric_index = metric.index(max_metric)
        history.append(data[metric_index])
        value, w = history[-1]

        if weight + w > 30:
            break

        weight += w
        history_extract.append(history[-1])
        data.remove((value, w))
        metric.remove(max_metric)
    
    extract_history(history_extract)
    

if __name__ == '__main__':
    main(data,history,history_extract,weight,go_out)
    
```
